```python
def mating_compatibility(male_type, female_type):
    if male_type == "A":
        if female_type == "A":
            mating_compatibility = A_males_preference_for_A_female*A_females_preference_for_A_male

        if female_type == "B":
            mating_compatibility = A_males_preference_for_B_female*B_females_preference_for_A_male
        
    if male_type == "B":

        if female_type == "A":
            mating_compatibility = B_males_preference_for_A_female*A_females_preference_for_B_male

        if female_type == "B":
            mating_compatibility = B_males_preference_for_B_female*B_females_preference_for_B_male

    return mating_compatibility
```
